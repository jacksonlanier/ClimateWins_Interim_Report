## .03 Scale Data

```python
# Create a subset that does not contain the date and month columns since this will not need to be scaled
df2 = df.drop(columns=['DATE', 'MONTH'])
```

```python
# Create scaler object using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
# Create new df with scaled data
df_scaled = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
df_scaled.head()
```

```python
# Insert the 'DATE' column at the beginning (position 0)
df_scaled.insert(0, 'DATE', df['DATE'])

# Insert the 'month' column at the beginning (position 1)
df_scaled.insert(1, 'MONTH', df['MONTH'])
```

```python
df_scaled.head()
```

```python
# Export scaled data
df_scaled.to_pickle(os.path.join(path, 'weather_scaled.pkl'))
```
